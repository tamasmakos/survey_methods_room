```python
import pandas as pd
enrich_address = pd.read_csv('enriched_address.csv')
```

```python
list(enrich_address.columns)
```

```python
# Merge pub, 'pub;cafe','pub;lottery'
enrich_address['pub'] = enrich_address['pub'] + enrich_address['pub;cafe'] + enrich_address['pub;lottery']
```

```python
# Drop 'bank;arts_centre'
enrich_address.drop('bank;arts_centre', axis=1, inplace=True)
```

```python
# Dropping columns that are not needed
columns_to_drop = ['id', 'town_ksh_code', 'province_name', 'town_name', 'town_type1', 'town_type2', 'town_type3', 'district_ksh_code', 'street_name', 'house_number', 'address', 'lon', 'lat',]

# Make a copy of the DataFrame
train = enrich_address.copy()
```

```python
train.drop(columns_to_drop, axis=1, inplace=True)
```

```python
# Showing all rows and columns
pd.set_option('display.max_rows', None)
description = train.describe().transpose()
```

```python
# Dropping columns where max value is 1
for col in description[description['max'] == 1].index:
    train.drop(col, axis=1, inplace=True)
```

```python
description = train.describe().transpose().sort_values('mean', ascending=False)
description
```

```python
# Dropping bench, waste_basket, ice_cream;cafe, yes
train.drop(['bench', 'waste_basket', 'ice_cream;cafe', 'yes'], axis=1, inplace=True)
```
