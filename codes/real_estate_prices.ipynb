```python
import os
import pandas as pd
import logging

# Set up logging
logging.basicConfig(filename='log.txt', level=logging.ERROR, format='%(asctime)s %(levelname)s %(message)s')

# Create an empty dataframe to store the data
prices = pd.DataFrame()

# Loop through all xlsx files in the directory
for file in os.listdir('/Users/tamasmakos/dev/survey_methods_room/data_files/real_estate_pices'):
    if file.endswith('.xlsx'):
        try:
            # Read the xlsx file
            xlsx = pd.read_excel('/Users/tamasmakos/dev/survey_methods_room/data_files/real_estate_pices/' + file, sheet_name=None, skiprows=1, header=[0,1,2], engine='openpyxl')
            
            # List to store dataframes from each sheet
            prices_list = []
            
            for name in xlsx.keys():
                if name != 'Tartalom':
                    # Clean up the headers to have a uniform structure
                    sheet_prices = xlsx[name]
                    sheet_prices.columns = ['_'.join(col).strip() for col in sheet_prices.columns.values]
                    
                    # Add the cleaned dataframe to the list
                    prices_list.append(sheet_prices)
            
            # Concatenate the cleaned sheets into one dataframe with axis=0
            prices_temp = pd.concat(prices_list, axis=0)
            
            # Add a column with the name of the file
            prices_temp['file_name'] = file
            
            # Append the dataframe to the main dataframe with axis=0
            prices = pd.concat([prices, prices_temp], axis=0)
        except Exception as e:
            # Log the error
            logging.error(f"Error processing file {file}: {e}")
            continue

# Display the first few rows of the dataframe
prices.head()
```

```python
# Selecting specific columns 0,1 and 8
prices = prices.iloc[:, [0,1,8,11]]
prices.head()
```

```python
# Renaming columns
prices.columns = ['town', 'street', 'average_price/sqm', 'year']
```

```python
# Selecting the year from year with a regex
prices['year'] = prices['year'].str.extract(r'(\d{4})')
```

```python
# Split the 'town' column into a new column 'town_raw'
prices['town_raw'] = prices['town'].str.split(' ')

# Extract the first and second elements from 'town_raw' to 'town' and 'distict' columns
prices['town'] = prices['town_raw'].str.get(0)
prices['distict'] = prices['town_raw'].str.get(1)

# Fill NaN values with empty strings to avoid issues during concatenation
prices['town'] = prices['town'].str.lower().fillna('')
prices['street_name'] = prices['street'].str.lower().fillna('')
prices['distict'] = prices['distict'].str.lower().fillna('')

# Create the 'address' column
prices['address'] = prices['town'] + ', ' + prices['distict'] + 'kerület ' + prices['street_name']
```

```python
# Dropping town_raw, distict and street_name columns
prices = prices.drop(['town_raw', 'street_name'], axis=1)
```

```python
prices.head()
```

```python
# Year to numeric
prices['year'] = pd.to_numeric(prices['year'])
```

```python
# Group by 'street' and 'year' and calculate the mean price for each street for each year
grouped = prices.groupby(['street', 'year'])['average_price/sqm'].mean().unstack()

# Calculate the change in price for each street between the years of interest
grouped['2016-2019_change'] = grouped[2019] - grouped[2016]
grouped['2009-2019_change'] = grouped[2019] - grouped[2009]

# Merge the calculated values back to the original DataFrame
prices = prices.merge(grouped[['2016-2019_change', '2009-2019_change', 2019]], on='street', how='left')
prices.rename(columns={2019: '2019_price'}, inplace=True)
```

```python
# Needs geocoding
needs_geo = merged_prices[merged_prices[['2016-2019_change', '2009-2019_change', '2019_price']].isna().all(axis=1)]
```

```python
prices_needs_geo = prices[prices['address'].isin(needs_geo['address'].unique())]
print(prices_needs_geo.shape[0])
prices_needs_geo.head()
```

```python
# Geocoding only the towns in the missing_towns list
import requests
def geocode(address):
    url = 'https://geocode.arcgis.com/arcgis/rest/services/World/GeocodeServer/findAddressCandidates'
    params = {'f': 'json', 'singleLine': address, 'outFields': 'Match_addr,Addr_type'}
    r = requests.get(url, params=params)
    if r.json()['candidates']:
        return r.json()['candidates'][0]['location']
    else:
        return None
    
prices_needs_geo['geo_code'] = prices_needs_geo['address'].apply(geocode)
```
